**Face Detection using Opencv**

```python
image_dir = "C:/Users/dhars/Downloads/Dhass/codeing/GUVI/2. MainBoot/4.Project_Code/Final Project - 1/Project6_Code/Intelliguard_PPE_Detection/4_authentication/face_Recognition/known_faces_images"

# Load Haar Cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# Loop through subfolders and images
for root, dirs, files in os.walk(image_dir):
    
    for file in files:
        if file.lower().endswith((".jpg", ".jpeg", ".png")):
            image_path = os.path.join(root, file)
            img = cv2.imread(image_path)

            if img is None:
                print(f" Could not read image: {image_path}")
                continue

            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

            if len(faces) == 0:
                print(f" No face detected in {file}")
            else:
                print(f" Detected {len(faces)} face(s) in {file}")
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)  
# (x, y) -> top-left corner of the rectangle.
# (x + w, y + h) ->  bottom-right corner of the rectangle.
# (0, 255, 0): the rectangle color  green (BGR format).
# 2: the thickness of the rectangle line in pixels.


cv2.imshow("Detected Face", img) #This opens a new window titled "Detected Face" and shows the image with the rectangle
#the window might open and close immediately.
#So use waitkey, it waits indefinitely for any key to be pressed.
cv2.waitKey(200) #make to 0 or make it to 20600seconds it wait for 20 sec and then it will automatically close
cv2.destroyAllWindows() #Closes the window after a close key is pressed.
```

```python
import matplotlib.pyplot as plt

for (x, y, w, h) in faces:
    face = img[y:y+h, x:x+w] #images in BGR format
    plt.imshow(cv2.cvtColor(face, cv2.COLOR_BGR2RGB)) #OpenCV loads images in BGR format.matplotlib expects RGB, so we convert using cv2.cvtColor
    plt.axis('off')
    plt.title("Detected Face")
    plt.show()
```
